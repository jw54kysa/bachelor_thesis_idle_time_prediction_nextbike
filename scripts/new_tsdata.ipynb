# Decompose TS Data
### in Trend, Seasonal, Resid

```python
import pandas as pd

tsdata = pd.read_csv('../data/time_series_data/ts_data_hex_881f1a8cb7fffff_3y.csv')
del tsdata['list']
del tsdata['bike_count']
tsdata = tsdata.fillna(0)
tsdata['date'] = pd.to_datetime(tsdata['date'])
tsdata = tsdata.set_index(['date']).asfreq('D')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
series = tsdata
result = seasonal_decompose(series, model='additive')
print(result.trend)
print(result.seasonal)
print(result.resid)
print(result.observed)
```

```python
from matplotlib import pyplot
fig = result.plot()
fig.set_size_inches((16, 9))
pyplot.show()
```
