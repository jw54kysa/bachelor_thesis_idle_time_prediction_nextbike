# Decompose TS Data
### in Trend, Seasonal, Resid

```python
import pandas as pd

# 1 hauptbahnhof
# 7 augustusplatz

tsdata = pd.read_csv('../../data/time_series_data/ts_data_881f1a8ca3fffff_Waldplatz.csv')
del tsdata['list']
del tsdata['bike_count']
tsdata = tsdata.fillna(0)
tsdata['date'] = pd.to_datetime(tsdata['date'])
tsdata = tsdata.set_index(['date']).asfreq('D')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

series = tsdata
result = seasonal_decompose(series, model='additive')
print(result.trend)
print(result.seasonal)
print(result.resid)
print(result.observed)
```

```python
from matplotlib import pyplot


fig = result.plot()
fig.set_size_inches((16, 9))
#pyplot.xlim(('2021-01-01','2021-02-01'))
pyplot.show()
```

```python
## Print Closeup of Seasonal for comparing to other hex


from matplotlib import pyplot as plt

fig, ax = plt.subplots()

plt.plot(result.seasonal['2021-01-01':'2021-02-01'])
ax.set_ylabel('Seasonal')
fig.set_size_inches(16, 1.5)
plt.show()
```
